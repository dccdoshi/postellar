```python
import pandas as pd 
import numpy as np 
import matplotlib.pylab as plt 
import torch
```

```python
df = pd.read_pickle('../data/SPIRou20_train.df')
df_sorted = df.sort_values(by=['teff', 'logg','metal']).reset_index(drop=True)

non_ones = np.where(df['Wavelength'].iloc[0] != 1)[0]
start = non_ones[0]
end = non_ones[-1]
```

```python
T_df = df_sorted[(df_sorted['logg'] == 5.0) & (df_sorted['metal'] == -0.5)]
plt.figure(figsize=(15,7))
for i in range(0,len(T_df),2):
    plt.plot(T_df['Wavelength'].iloc[i][start:end],T_df['Final'].iloc[i][start:end],label='T:'+str(T_df['teff'].iloc[i]))
plt.ylabel("Normalized Flux")
plt.xlabel("Wavelength (nm)")
plt.title("Trainning Spectra at different Temperatures with logg=5.0, Fe/H=-0.5")
plt.legend()
# plt.xlim(1515,1520)
```

```python
logg_df = df_sorted[(df_sorted['teff'] == 3300) & (df_sorted['metal'] == -0.5)]
plt.figure(figsize=(15,7))
for i in range(len(logg_df)):
    plt.plot(logg_df['Wavelength'].iloc[i][start:end],logg_df['Final'].iloc[i][start:end],label='logg:'+str(logg_df['logg'].iloc[i]))
plt.ylabel("Normalized Flux")
plt.xlabel("Wavelength (nm)")
plt.title("Trainning Spectra at different logg with teff=3300K, Fe/H=-0.5")
# plt.xlim(1515,1520)
plt.legend()
```

```python
metal_df = df_sorted[(df_sorted['teff'] == 3000) & (df_sorted['logg'] == 5.0)]
plt.figure(figsize=(15,7))
for i in range(len(metal_df)):
    plt.plot(metal_df['Wavelength'].iloc[i][start:end],metal_df['Final'].iloc[i][start:end],label='Fe/H:'+str(metal_df['metal'].iloc[i]))
plt.ylabel("Normalized Flux")
plt.xlabel("Wavelength (nm)")
plt.title("Trainning Spectra at different Fe/H with teff=3300K, logg=-5.0")
# plt.xlim(1520,1525)
plt.legend()
```
